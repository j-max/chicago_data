```python
pr_df [ 
    (pr_df["school_survey_ambitious"] == "VERY_STRONG") 
  & (pr_df["school_survey_collaborative"] == "VERY_STRONG")
  & (pr_df["school_survey_effective"] == "VERY_STRONG")
  & (pr_df["school_survey_involved"] == "VERY_STRONG")
  & (pr_df["school_survey_safety"] == "VERY_STRONG")
]

```

```python
map_survey_scores = {
    "NOT ENOUGH DATA": 0,
    "VERY WEAK": 1,
    "WEAK": 2,
    "NEUTRAL": 3,
    "STRONG": 4,
    "VERY STRONG": 5
}
```

```python
pr_df["score_school_survey_ambitious"] =  pr_df["school_survey_ambitious"].map(map_survey_scores)
pr_df["score_school_survey_effective"] =  pr_df["school_survey_effective"].map(map_survey_scores)
pr_df["score_school_survey_collaborative"] =  pr_df["school_survey_collaborative"].map(map_survey_scores)
pr_df["score_school_survey_involved"] =  pr_df["school_survey_involved"].map(map_survey_scores)
pr_df["score_school_survey_safety"] =  pr_df["school_survey_safety"].map(map_survey_scores)
```

```python
pr_df["pr_total_survey_score"] = (
    pr_df["score_school_survey_ambitious"]
    + pr_df["score_school_survey_effective"]
    + pr_df["score_school_survey_collaborative"] 
    + pr_df["score_school_survey_involved"]
    + pr_df["score_school_survey_safety"] 
)
```
